```python
import string as str
from requests import get
url='https://www.imdb.com/chart/top/?ref_=nv_mv_250'
response=get(url)
response.text
from bs4 import BeautifulSoup
soup_object=BeautifulSoup(response.text,'html.parser')
#look for tr tags
#However there might be other tables on that page having rows that contain something other than movie information
#narrowdown: tbody class="lister-list"> this is the one we want.

movie_table=soup_object.find('tbody',class_='lister-list')
                             
#this will take just the html from within the tbody tag with a class of lister list and save it. 

movie_list=movie_table.find_all('tr')

#this will take the html from within each tr

names=[]
years=[]
imdb_ratings=[]

for movie in movie_list:
    name=movie.find('td',class_='titleColumn').a.text
    names.append(name)
    year1=movie.find('span',class_='secondaryInfo').text
    years.append(year)
    rating=float(movie.find('td',class_='ratingColumn imdbRating').strong.text)
    imdb_ratings.append(rating)



```
